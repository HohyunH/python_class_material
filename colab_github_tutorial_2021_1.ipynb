## Github 주요 용어 정리

- 원격 저장소 : 파일이 원격 저장소 전용 서버에서 관리되며 여러 사람이 함께 공유
- 로컬 저장소 : 원격 저장소를 로컬로 다운받아서 관리함. 로컬에서 작업한 내용을 원격 저장소로 push할 수 있음.
- branch : 개발하는 과정에서 버전 관리를 하기 위한 도구. 최초에 저장소 생성시 master branch가 자동적으로 생성되며, 필요에 따라 develop branch, hot-fix branch등 만들어서 작업 시행. 이후 작업한 내용에 이상이 없으면 master branch으로 병합.
- merge : 분기해서 작업한 내용을 master branch로 병합하는 명령어


![branch 설명](https://github.com/HohyunH/git-practice/blob/main/%EA%B7%B8%EB%A6%BC1.png?raw=true)


- add : commit하기에 앞서 저장을 원하는 파일을 묶는 (스테이지에 파일을 올리는) 작업
- commit : 각 수정사항마다 어떤 내용이 변경되었는지 짧게 기록하는 메세지. 파일의 save개념(code 수정 후 다시 돌아올 수 있다)
- push : 로컬 저장소에서 commit한 내용을 원격 저장소로 보내는 과정
- pull : 로컬 저장소의 내용이 오래됐거나 원격 저장소의 내용이 변경되었을 경우, 최신버전으로 갱신해주는 명령어


![git 설명 그림](https://i.ytimg.com/vi/0nqJKEh3YCc/maxresdefault.jpg)




### Git Lifecycle

![Github 파일의 라이프 사이클](https://git-scm.com/book/en/v2/images/lifecycle.png)

Untracked 
- Git과 아무 상관이 없는 상태
- 따라서 Git이 대상 파일을 관리하지 못함
- 최초 add를 해줘야 Git의 관리 대상이 됨
- Git이 관리하는 파일을 삭제하면 Untracked가 됨

Unmodified
- 코드 저장이완료된 상태
- Staged 상태에서 commit을 하면 Unmodified가 됨

Modified
- Git으로 관리되고 있던 코드를 수정하여 변경이 일어난 상태
- commit 할수 없음(commit하려면 Staged 상태가 되어야 함)

Staged
- 이제 코드를 저장해도 좋다는 (commit이 가능한) 상태
- Untracked/Modified 상태의 파일은 add하면 Staged가 됨

## fork와 clone

- fork : fork는 다른 사람의 Github repository에서 내가 어떤 부분을 수정하거나 추가 기능을 넣고 싶을 때, 해당 repository를 내 Github repository로 그대로 복제하는 기능
![fork](https://github.com/HohyunH/git-practice/blob/main/fff.PNG?raw=true)

- clone : clone은 특정 repository를 내 local machine에 복사하여 새로운 저장소를 만드는 것
![fork and clone](https://media.vlpt.us/post-images/imacoolgirlyo/cbe5ca40-5f44-11e9-88b2-25d00148b532/gitfork.png)


example1) push 예제
```
$ mkdir test1
$ cd test1
$ git init # 저장소가 Git 저장소가 되었다고 설정
$ git clone https://github.com/HohyunH/git-practice.git
$ git add main.py # 파일을 stage로 올려주는(더해주는) 방법.
$ git commit -m "main function" # commit message는 항상 친절하게 적어준다.
$ git push origin master # origin : 원격 저장소 브랜치, master : 로컬 저장소 브랜치
```
example2) pull 예제
```
$ mkdir test2
$ cd test2
$ git init
$ git clone https://github.com/testpage/push_example.git
$ git pull origin(원격저장소 브랜치) master(로컬저장소 브랜치)
```
